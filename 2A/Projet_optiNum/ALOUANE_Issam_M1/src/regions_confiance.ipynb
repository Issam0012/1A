3. Quels sont les avantages et inconvénients des deux approches ?

Pour le pas de Cauchy, elle converge rapidement en une seule itération (quand c'est possible), tandis que pour le gradient conjugué tronqué, la convergence est lente mais donne des résultats plus précises avec une décroissance qui peut être négatif (selon le cas), ce qui n'est pas possible pour le pas de Cauchy
